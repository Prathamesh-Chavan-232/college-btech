```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as mt
import plotly
import plotly.express as px

data = pd.read_csv('Iris.csv',sep=',')
data.head()
```

```python
data.shape
```

```python
# 2d Scatter plot using matplotlib
data.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
mt.show()
```

```python
# 2d Scatter plot using seaborn
sns.set_style("whitegrid")
sns.FacetGrid(data,hue='Species').map(mt.scatter,"SepalLengthCm","SepalWidthCm").add_legend()
mt.show()
```

```python
# 3d scatter using plotly 

iris = px.data.iris()
iris.head()
```

```python

fig = px.scatter_3d(iris,x='sepal_length',y='sepal_width',z='species',color='petal_width',size='petal_length')
fig.show()
```

```python
sns.violinplot(x='species',y='petal_length',data=iris,size=8)
mt.show()
```
